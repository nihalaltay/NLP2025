```python
#Cleaning the data part 1: Remove text between brackets [ ] (The song names).
#We cannot do this after the tokenization because it turns into a list

lyrics_adamlar = re.sub(r'\[.*?\]', '', lyrics_adamlar)
lyrics_cte = re.sub(r'\[.*?\]', '', lyrics_cte)

#print(lyrics_adamlar)
#print(lyrics_cte)
```

```python
#Tokenizing

import nltk
nltk.download('punkt_tab')
from nltk.tokenize import word_tokenize

adamlar_tokenized = word_tokenize(lyrics_adamlar)
#print(adamlar_tokenized)

cte_tokenized = word_tokenize(lyrics_cte)
#print(cte_tokenized)
```

```python
#Cleaning the data part 2: Cleaning the data for punctuation
#We are doing this after tokenization to keep words like "I'm" as seperate tokens

def clean_tokens(tokens):

    cleaned_tokens = []
    for token in tokens:

        #Remove punctuation
        token = re.sub(r'[^\w\s]', '', token)

        #Convert lyrics to lowercase
        token = token.lower()

        #Add tokens that are not empty to the list
        if token.strip():
            cleaned_tokens.append(token)

    return cleaned_tokens

clean_adamlar = clean_tokens(adamlar_tokenized)
# print(clean_adamlar)

clean_cte = clean_tokens(cte_tokenized)
# print(clean_cte)
```

```python
#Calculate the word lenghts and frequencies

import pandas as pd

#For Adamlar

#Create a dataframe from the list
df_adamlar = pd.DataFrame(clean_adamlar, columns=['word'])

#Calculate frequency using value_counts and map it back to a column
df_adamlar['frequency'] = df_adamlar['word'].map(df_adamlar['word'].value_counts())

#Calculate word lengths
df_adamlar['length'] = df_adamlar['word'].apply(len)

#Remove duplicate rows for a summary table
df_adamlar = df_adamlar.drop_duplicates().reset_index(drop=True)

print(df_adamlar)


#For CTE

#Create a dataframe from the list
df_cte = pd.DataFrame(clean_cte, columns=['word'])

#Calculate frequency using value_counts and map it back to a column
df_cte['frequency'] = df_cte['word'].map(df_cte['word'].value_counts())

#Calculate word lengths
df_cte['length'] = df_cte['word'].apply(len)

#Remove duplicate rows for a summary table
df_cte = df_cte.drop_duplicates().reset_index(drop=True)

print(df_cte)
```

```python
#Visualization

import matplotlib.pyplot as plt
import seaborn as sns

#Create two subplots side-by-side for the two visualizations
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

#Scatter plot for Adamlar
sns.scatterplot(ax=axes[0], data=df_adamlar, x="length", y="frequency", s=100)
axes[0].set_title("Adamlar: Frequency vs. Length")
axes[0].set_xlabel("Word Length (Characters)")
axes[0].set_ylabel("Frequency")

#Add word labels to each point
for i, row in df_adamlar.iterrows():
    axes[0].text(row['length'] + 0.1, row['frequency'] + 0.1, row['word'], fontsize=9)

#Scatter plot for Cage the Elephant (CTE)
sns.scatterplot(ax=axes[1], data=df_cte, x="length", y="frequency", s=100)
axes[1].set_title("Cage the Elephant: Frequency vs. Length")
axes[1].set_xlabel("Word Length (Characters)")
axes[1].set_ylabel("Frequency")

#Add word labels to each point
for i, row in df_cte.iterrows():
    axes[1].text(row['length'] + 0.1, row['frequency'] + 0.1, row['word'], fontsize=9)

#Adjust layout to prevent overlap and display the plot
plt.tight_layout()
plt.show()
```
